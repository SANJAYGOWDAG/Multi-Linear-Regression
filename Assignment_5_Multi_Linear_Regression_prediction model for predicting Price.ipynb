# Data preparation

```python
cars_data=corolla_cars_data.loc[:,["Price","Age_08_04","KM","HP","cc","Doors","Gears","Quarterly_Tax","Weight"]]
```

```python
cars_data.head()
```

```python
corr_matrix=cars_data.corr()
```

```python
plt.figure(figsize=[12,10])
sns.heatmap(corr_matrix,annot=True)
```

```python
#### there is no multicoliniarity problem  with in the independent features

```

```python
x=cars_data.iloc[:,1:]
y=cars_data.iloc[:,0]
```

```python
cars_data.head()
```

```python
x.head()
```

```python
x['Gears'].value_counts() ,x['Doors'].value_counts()
```

```python
y.head()
```

```python
# splittinig the dataset into training set and Test set
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=34)
```

# Model building and Model training

```python
from sklearn.linear_model import LinearRegression
model_1=LinearRegression() # model initialization
```

```python
# fitting multiple linear regression model to the training set
model_1.fit(x_train,y_train)
```

```python
# predicting the Test results
```

```python
y_pred=model_1.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
r2_score_m1=r2_score(y_test,y_pred).round(3)
r2_score_m1
```

```python
MSE=mean_squared_error(y_test,y_pred).round(3)
MSE
```

```python
# all the features are in different scales so we have to standerdize the input features 
from sklearn.preprocessing import StandardScaler
std_scaler=StandardScaler()
scaled_x=std_scaler.fit_transform(x)
scaled_x=pd.DataFrame(scaled_x,columns=['Age_08_04','KM','HP','cc','Doors','Gears','Quarterly_Tax','Weight'])
```

```python
scaled_x
```

```python
scaled_x_train,scaled_x_test,y_train,y_test=train_test_split(scaled_x,y,test_size=0.2,random_state=23)
```

# model building with transformed independent feature

```python
model_2=LinearRegression()
model_2.fit(scaled_x_train,y_train)
```

```python
y_pred_2=model_2.predict(scaled_x_test)
```

```python
r2_score_m2=r2_score(y_test,y_pred_2).round(4)
r2_score_m2
```
